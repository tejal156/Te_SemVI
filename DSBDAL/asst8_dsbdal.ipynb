# handling null values

```python
data.isna().sum()
```

```python
data.info()
```

```python
data["CRIM"].fillna( data["CRIM"].mode()[0] , inplace = True)
data["ZN"].fillna( data["ZN"].mode()[0] , inplace = True)
data["INDUS"].fillna( data["INDUS"].mode()[0] , inplace = True)
data["CHAS"].fillna( data["CHAS"].mode()[0] , inplace = True)
data["AGE"].fillna( data["AGE"].mode()[0] , inplace = True)
data["LSTAT"].fillna( data["LSTAT"].mode()[0] , inplace = True)
```

```python
data.isna().sum()
```

# train test split

```python
data
```

```python
# es, machine learning models can handle floating-point numeric values. In fact, many datasets, including the Boston housing dataset, consist primarily of floating-point numeric values.
```

```python
X = data.iloc[ : , 0 : -1]
```

```python
Y = data.iloc[: , -1 ]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split( X , Y , test_size = 0.2 , random_state = 42)
```

# MODEL training

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train , y_train )
```

```python
y_test_pred = lr.predict(x_test)
```

```python
y_train_pred = lr.predict(x_train)
```

# Evaluating model

```python
lr.score( x_test , y_test)
```

```python
lr.score(x_train , y_train)
```

```python
from sklearn.metrics import mean_squared_error , r2_score
```

```python
import math
math.sqrt( mean_squared_error( y_test , y_test_pred))
```

```python
mean_squared_error( y_train , y_train_pred)
```

```python
r2_score( y_test , y_test_pred)
```

```python
r2_score( y_train , y_train_pred)
```
